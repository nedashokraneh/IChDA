```python
structural_signals_path = "../../data/GM12878/hg19/res100000_datasets/structural_signals.txt"
structural_signals = np.loadtxt(structural_signals_path)

functional_signals_path = "../../data/GM12878/hg19/res100000_datasets/signals.txt"
functional_signals = np.arcsinh(np.loadtxt(functional_signals_path))

combined_signals = np.concatenate([functional_signals, structural_signals], axis = 1)

signals_df = pd.DataFrame(functional_signals)
signals_df.columns = signals_names

#signals_df = pd.concat([signals_df,RT_df[['G1', 'S1', 'S2', 'S3', 'S4', 'G2']]], axis = 1)

k = 6
lengths = GM12878_dataset.get_chunks_lengths()

func_hmm = hmm.GaussianHMM(n_components=k, covariance_type = 'full', algorithm='viterbi')
func_hmm.fit(functional_signals, lengths)
func_l = func_hmm.predict(functional_signals, lengths)

struct_hmm = hmm.GaussianHMM(n_components=k, covariance_type = 'full', algorithm='viterbi')
struct_hmm.fit(structural_signals, lengths)
struct_l = struct_hmm.predict(structural_signals, lengths)

comb_hmm = hmm.GaussianHMM(n_components=k, covariance_type = 'full', algorithm='viterbi')
comb_hmm.fit(combined_signals, lengths)
comb_l = comb_hmm.predict(combined_signals, lengths)

half_func_hmm = hmm.GaussianHMM(n_components=k, covariance_type = 'full', algorithm='viterbi')
half_func_hmm.fit(functional_signals[:,[0,2,4,6,8,10]], lengths)
half_func_l = half_func_hmm.predict(functional_signals[:,[0,2,4,6,8,10]], lengths)
```

```python
func_entropies = entr(np.round(func_hmm.predict_proba(functional_signals),2)).sum(axis = 1)
struct_entropies = entr(np.round(struct_hmm.predict_proba(structural_signals),2)).sum(axis = 1)
comb_entropies = entr(np.round(comb_hmm.predict_proba(combined_signals),2)).sum(axis = 1)
half_func_entropies = entr(np.round(half_func_hmm.predict_proba(functional_signals[:,[0,2,4,6,8,10]]),2)).sum(axis = 1)
```

```python
print('average entropy of HMM_functional annotations: {}, HMM_structural annotations: {}, and HMM_combined annotations: {}.'.format(np.mean(func_entropies), np.mean(struct_entropies), np.mean(comb_entropies)))
```

```python
def gene_expression_ve(annotation_path, annotation_resolution, gene_expression_path):
    annotation_dict = evaluation_utils.make_annotation_dict(annotation_path, annotation_resolution)
    expressions_dict = {}
    labels_dict = {}
    expressions = []
    labels = []
    with open(gene_expression_path, 'r') as f:
        for line in f:
            gene_id, chr_num, pos1, pos2, _, gene_expr = line.rstrip("\n").split("\t")
            if not chr_num in [str(c) for c in np.arange(1,23)]:
                continue
            chr_name = 'chr{}'.format(chr_num)
            start = min(int(pos1),int(pos2))
            end = max(int(pos1),int(pos2))
            label = evaluation_utils.get_most_freq_label(chr_name, start, end, annotation_dict, annotation_resolution)
            if label == None:
                continue
            if chr_name in expressions_dict:
                expressions_dict[chr_name].append(float(gene_expr))
                labels_dict[chr_name].append(label)
            else:
                expressions_dict[chr_name] = [float(gene_expr)]
                labels_dict[chr_name] = [label]
            expressions.append(float(gene_expr))
            labels.append(label)
    VE = evaluation_utils.variance_explained(np.arcsinh(expressions),labels)
    chr_VEs = {}
    for key in expressions_dict:
        chr_VEs[key] = evaluation_utils.variance_explained(np.arcsinh(expressions_dict[key]), labels_dict[key])
    return chr_VEs
```

```python
def RT_score(RT_path, annotation_path, annotation_resolution):    
    annotation_dict = evaluation_utils.make_annotation_dict(paths['HMM_functional'], annotation_resolution)
    RT = pd.read_csv(RT_path, sep = "\t", header = None)
    RT.columns = ['chr_name', 'pos', 'G1', 'S1', 'S2', 'S3', 'S4', 'G2']
    RT['label'] = [annotation_dict[chr_name][int(pos)] for chr_name, pos in zip(RT['chr_name'].values, RT['pos'].values/annotation_resolution)]
    RT = RT[RT['label'].values != None]
    chr_VEs = {}
    chroms = np.unique(RT['chr_name'])
    for chrom in chroms:
        curr_chr_VEs = []
        for phase in ['G1', 'S1', 'S2', 'S3', 'S4', 'G2']:
            ve = evaluation_utils.variance_explained(RT[RT['chr_name'] == chrom][phase], 
                                                     RT[RT['chr_name'] == chrom]['label'])
            curr_chr_VEs.append(ve)
        chr_VEs[chrom] = np.mean(curr_chr_VEs)
    return chr_VEs
#RT['RT_label'] = RT.iloc[:,2:8].to_numpy().argmax(axis=1)
#VEs = []
#for phase in ['G1', 'S1', 'S2', 'S3', 'S4', 'G2']:
#    VEs.append(variance_explained(RT[phase],RT['label']))
#mean_VE = np.mean(VEs)
#ARI = metrics.adjusted_rand_score(RT['RT_label'],RT['label'])
```

```python
def create_pivot_table_per_chrom_from_bedpe(bedpe_path, annotation_path, annotation_resolution):

    annotation_dict = evaluation_utils.make_annotation_dict(annotation_path, annotation_resolution)
    labels1 = {}
    labels2 = {}
    weights = {}
    with open(bedpe_path, 'r') as f:
        for line in f:
            chr1, start1, end1, chr2, start2, end2, weight = line.split("\t")
            if not (evaluation_utils.is_valid_chr(chr1)) & (evaluation_utils.is_valid_chr(chr2)):
                continue
            if chr1 != chr2:
                continue
            if chr1 in labels1:
                labels1[chr1].append(evaluation_utils.get_most_freq_label(chr1, int(start1), int(end1), annotation_dict, annotation_resolution))
                labels2[chr1].append(evaluation_utils.get_most_freq_label(chr2, int(start2), int(end2), annotation_dict, annotation_resolution))
                weights[chr1].append(int(weight))
            else:
                labels1[chr1] = [evaluation_utils.get_most_freq_label(chr1, int(start1), int(end1), annotation_dict, annotation_resolution)]
                labels2[chr1] = [evaluation_utils.get_most_freq_label(chr2, int(start2), int(end2), annotation_dict, annotation_resolution)]
                weights[chr1] = [int(weight)]
    contacts = {}
    for key in labels1.keys():
        contacts[key] = pd.DataFrame({'label1': labels1[key], 'label2': labels2[key], 
                                      'weight': weights[key]}).groupby(['label1','label2'])['weight'].sum().reset_index()
        contacts[key] = contacts[key].pivot_table(columns = 'label2', index = 'label1', values = 'weight')
        contacts[key] = contacts[key].fillna(0)
    return contacts

def get_coverage_per_chrom(annotation_path):

    coverages = {}
    total = {}
    with open(annotation_path, 'r') as f:
        for line in f:
            chr_name, start, end, label, *_ = line.rstrip("\n").split("\t")
            length = int(end) - int(start)
            if label == 'NA':
                continue
            if chr_name in coverages:
                if label in coverages[chr_name]:
                    coverages[chr_name][label] = coverages[chr_name][label] + length
                else:
                    coverages[chr_name][label] = length
                total[chr_name] = total[chr_name] + length
            else:
                coverages[chr_name] = {}
                coverages[chr_name][label] = length
                total[chr_name] = length
    for key in coverages:
        for label in coverages[key]:
            coverages[key][label] = coverages[key][label] / total[key]
    #sns.barplot(x = freq.index.values, y = freq.values)
    return coverages

def OE_ratio(contacts, coverages):
    total_contacts = np.sum(contacts.to_numpy())
    expected_intra_contacts = 0
    observed_intra_contacts = 0
    for key in coverages:
        expected_intra_contacts = expected_intra_contacts + coverages[key]**2
        observed_intra_contacts = observed_intra_contacts + contacts.loc[key,key]/total_contacts
    return observed_intra_contacts/expected_intra_contacts

def OE_ratio_per_chrom(all_contacts, all_coverages):
    OE_ratios = {}
    for key in all_contacts.keys():
        OE_ratios[key] = OE_ratio(all_contacts[key], all_coverages[key])
    return OE_ratios


def OE_intra_of_bedpe_per_chrom(bedpe_path, annotation_path, annotation_resolution):
    contacts = create_pivot_table_per_chrom_from_bedpe(bedpe_path, annotation_path, annotation_resolution)
    coverages = get_coverage_per_chrom(annotation_path)
    OE_ratios = OE_ratio_per_chrom(contacts, coverages)
    return OE_ratios
```
