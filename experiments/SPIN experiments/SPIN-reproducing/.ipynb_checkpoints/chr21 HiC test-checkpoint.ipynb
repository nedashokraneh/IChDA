```python
### data ###
input_signals = pd.read_csv("data/K562_TSA_DamID_25k.no_NAN.bed", sep = "\t", header = None)

bins = input_signals.iloc[:,0:3]
bins['index'] = np.arange(0,bins.shape[0])
bins.to_csv("data/bin.txt", sep = "\t", index = False, header = None)

inputs = input_signals.iloc[:,3:]
inputs.to_csv("data/input.txt", sep = "\t", index = False, header = None)
```

```python
bin_map = data_utils.get_chr_pos_to_index2("data/bin.txt",25000,"hg38")
```

```python
p_value_dir = "../../../data/K562/Hi-C/hg38/oe_25000_weibull/"
hic_data = pd.read_csv(os.path.join(p_value_dir, 'chr1_chr1.txt'), sep = "\t", header = None)
```

```python
f = hic_data[hic_data.iloc[:,3]<0.00001]
f.columns = ['first_pos', 'second_pos', 'weight', 'pvalue']
f.loc[:,'first_index'] = [bin_map['chr1'][int(p/25000)] for p in f.loc[:,'first_pos']]
f.loc[:,'second_index'] = [bin_map['chr1'][int(p/25000)] for p in f.loc[:,'second_pos']]
f = f.dropna()
f = f.loc[:,['first_index','second_index']].astype(int).sort_values(['first_index','second_index'])
```

```python
s = max(f['first_index'].max(),f['second_index'].max())
```

```python
f_mat = np.zeros((s+1,s+1))
for i,row in f.iterrows():
    f_mat[row['first_index'], row['second_index']] = 1
    f_mat[row['second_index'], row['first_index']] = 1
```

```python
sns.heatmap(f_mat[0:100,0:100])
```
