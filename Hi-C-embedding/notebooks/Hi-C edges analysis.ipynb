```python
###### calculating density of all intra-chromosomal interactions ######
chr_intra_dense = []
for chr_num in np.arange(1,23):
    interactions_list, _ = utils.create_hic_list('GM12878',chr_num,chr_num,100000,'oe')
    observed_edges = interactions_list.shape[0]
    valid_nodes = len(np.unique(interactions_list['source']))
    expected_edges = math.pow(valid_nodes,2)
    chr_intra_dense.append({'chr': 'chr'+str(chr_num), 'density': observed_edges/expected_edges})
chr_intra_dense.to_csv("supp/intra_chrom_dense.txt", index = False, sep = "\t")
###### all are around 0.46-0.5 ######
```
